```python
#5) Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.
```

```python
import numpy as np
import pandas as pd
np.random.seed(0)
num_samples = 100
num_features = 5
data = np.random.randn(num_samples, num_features)
corr_matrix = np.array([
    [1.0, 0.8, 0.6, 0.4, 0.2],
    [0.8, 1.0, 0.7, 0.5, 0.3],
    [0.6, 0.7, 1.0, 0.6, 0.4],
    [0.4, 0.5, 0.6, 1.0, 0.5],
    [0.2, 0.3, 0.4, 0.5, 1.0]
])
cov_matrix = np.linalg.cholesky(corr_matrix) @ np.linalg.cholesky(corr_matrix).T
correlated_data = np.dot(data, cov_matrix)
df = pd.DataFrame(correlated_data, columns=[f'Feature_{i+1}' for i in range(num_features)])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
corr_matrix_df = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix_df, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
np.random.seed(30)
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}
df = pd.DataFrame(data)
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['X'], df['Y'], df['Z'], c='b', marker='o')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('3D Scatter Plot of Data Points')
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
np.random.seed(15)
data_grades = {
    'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
    'Score': np.random.randint(50, 100, 200)
}
df_grades = pd.DataFrame(data_grades)
```
